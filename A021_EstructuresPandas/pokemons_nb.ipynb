## Màscares de substitució.

### Fes que tots els pokemon amb HP < 50, que ara tinguin HP = 50.


### Fes que tots els pokemon de generació 1 tinguin 20 punts més d'HP.


```python
pokedex2 = copy.deepcopy(pokedex)

# Creem màscara.
mask2 = (pokedex2.loc[ : , 'HP' ] <= 50)

# Pas 2. Substituim valors màscara.
pokedex2.loc[ mask2, ['HP'] ] = 50 

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2.loc[mask2, : ])
```

```python
pokedex2 = copy.deepcopy(pokedex)

# Creem màscara.
mask2 = (pokedex2.loc[ : , 'Generation' ] == 1)

# Pas 2. Substituim valors màscara.
pokedex2.loc[ mask2, ['HP'] ] += 20 

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2.loc[mask2, : ])
```


### Fes que tots els pokemon de tipus elèctric, que en vez de dirse 'Electric', ahora tendran un símbol d'un raig. 🗲 => U+1F5F2

```python
pokedex2 = copy.deepcopy(pokedex)

# Màscara de substitució de valors.

# Pas 1. Aplicar la màscara per seleccionar els que són de tipus elèctric.
mask_electrics = (pokedex2.loc[ : , 'Type 1' ] == 'Electric') | (pokedex2.loc[ : , 'Type 2' ] == 'Electric')

# Pas 2. Aplicar la màscara; és a dir, modificar només les files del dataframe que compleixen la màscara.
#  df. 🗲 => U+1F5F2
pokedex2.loc[mask_electrics,['Type 1'] ] = "🗲"
pokedex2.loc[mask_electrics,['Type 2'] ] = "🗲"

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2.loc[mask_electrics, : ])
```


### Fes que els valors np.nan que hi ha a la filera de Type 2 hi posi "Not Avaliable"
Ref:
https://pandas.pydata.org/docs/user_guide/10min.html#missing-data

Pista: mètode fillna
fillna(value="Not Avaliable")

```python
pokedex2 = copy.deepcopy(pokedex)

# Primer mirem si hi ha han.
print("Nan values", pokedex2['Type 2'].isnull().values.sum())

# Pas. Podem aplicar dirèctament la màscara, fillna el que fa és 
# mirar si és np.nan o no, i si ho hes ho canvia pel valor.
mask_change_nan = pokedex2.loc[:,'Type 2'].fillna('Not Avaliable')
pokedex2.loc[:,'Type 2'] = mask_change_nan

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2)
```

```python
### Solució més curta, en aquest cas que només tenim NaN en una fila.
pokedex2 = copy.deepcopy(pokedex)

# Tot el fitxer
pokedex2.fillna(value = "Not Available")

# Només la columna
pokedex2.loc[:,'Type 2'].fillna(value = "Not Available")
```
