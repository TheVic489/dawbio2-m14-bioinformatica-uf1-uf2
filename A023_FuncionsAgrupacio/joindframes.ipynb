## Exemples GROUPBY

https://pandas.pydata.org/pandas-docs/stable/user_guide/10min.html#grouping

```python
df_demo_group = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo", "biz", "biz"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three", "four", "three"],
        "C": np.random.randn(10),
        "D": np.random.randn(10),
    }
)
df_demo_group
```

```python
# group by a colum(or more)
df_demo_group.groupby("A")[["C", "D"]].sum()
```

```python
# group by a colum(or more)
df_demo_group.groupby("A")[["C", "D"]].count()
```

```python
df_demo_group.groupby(["A", "B"]).sum()
```
