### Crea un dataframe amb dades de 100 pacients generades aleatòriament.
Per ara ens interessen la edat, el pes i l'altura.

```python
import numpy as np 
import pandas as pd
import copy

min_value_edat = 20
max_value_edat = 120
min_value_altura = 1.20
max_value_altura = 2.20
min_value_pes = 40
max_value_pes = 140

df_pacients = pd.DataFrame({ 
    'edat' : np.random.randint(min_value_edat, max_value_edat ,100),
    'altura' : np.random.uniform(min_value_altura, max_value_altura , size=100),
    'pes' : np.random.randint(min_value_pes, max_value_pes, size=100)
})
df_pacients

## Copia de seguretat.
```

#### Crea una nova columna, imc; en la que el seu valor sigui el resultat de l'índex de massa corporal.
imc = pes / (altura ** 2)

Referència:
https://python-forum.io/thread-18759.html

```python
def calcula_imc(df_pacients) -> float:
    return round(df_pacients['pes'] / df_pacients['altura'] ** 2, 4)

# axis=1, apliquem resultats a les files.
df_pacients['imc'] = df_pacients.apply(calcula_imc, axis=1)

#dfcopy.apply(lambda column:column.sum())
df_pacients
```

#### Mostra la mitjana aritmètica de l'IMC de tots els pacients.

```python
df_pacients["imc"].mean()
```
