### En el meu cas, vull convertir a camp categòric el grup d'edat.
Recurs:
https://github.com/miquelamorosaldev/dawbio2-m14-bioinformatica-uf1-uf2/blob/main/A023_FuncionsAgrupacio/AutoCategoricalGrade.ipynb

````python
# uso aquesta consulta per mirar quins valors pot tenir el categorical, si hi ha molts.
grups_edat2 = df_seqvariants_covid.drop_duplicates(subset=['grup_edat'])
display(df_seqvariants_covid.groupby('grup_edat')['grup_edat'].count())
````

````python
df_seqvariants_covid_orig = copy.deepcopy(df_seqvariants_covid)

# 1 - Definim l'ordre i els valors dels categòrics.
grups_edat = ['0','1 i 2','3 i 4','15 a 44','45 a 59','60 a 69','70 a 79','80 o més','No disponible']

# 2 - Si tingués edats concretes i les volgués posar en un grup hauria de crear una funció.
## grups_edat: list[str] = ["Insuficiente", "Suficiente", "Bien", "Notable", "Excelente"]

# 3 - Convertir a Category dtype
df_seqvariants_covid_orig.loc[:, "grup_edat"] = df_seqvariants_covid_orig.loc[:, "grup_edat"].astype("category")

# 4 - Assignar el valor de la categoria.
df_seqvariants_covid_orig.loc[:, "grup_edat"] = df_seqvariants_covid_orig.loc[:, "grup_edat"].cat.set_categories(grups_edat)

print(df_seqvariants_covid_orig.groupby('grup_edat')['grup_edat'].count())

# 5 - Provem que funciona
print(df_seqvariants_covid_orig.dtypes)
# df_seqvariants_covid.shape

display(df_seqvariants_covid_orig.sort_values(by=["grup_edat"],ascending=True,inplace=True))
````

#### Per a crear un gràfic de categories he fet un truc, convertir-lo en una sèrie (value_counts).

````python
df_seqvariants_covid_orig['grup_edat'].value_counts().sort_index().plot(kind='bar', xlabel='grup_edat', ylabel='Count')
````
