```python
import utils
import pandas as pd

entries_df: pd.DataFrame = utils.read_csv_file("scimago-medicine.csv")

# Select columns, group, calculate max, sort by HIndex
region_max_hindex: pd.DataFrame = ( entries_df.loc[:, ["Region", "H index"]]
        .groupby("Region")
        .max()
        .sort_values(by="H index", ascending=False)
)

display(region_max_hindex)

# Plot result
region_max_hindex.plot(kind="bar").get_figure().savefig("s10.pdf")
```
